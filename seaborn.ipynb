```python
import pandas as pd
import seaborn as sns
```

# Saved CSVs

```python
bare2048 = pd.read_csv("./Outputs/CSV/bare.csv")
bare1024 = pd.read_csv("./Outputs/CSV/1024bare.csv")
bare1024lr01 = pd.read_csv("./Outputs/CSV/1024bare.01lr.csv")
bare1024lr1 = pd.read_csv("./Outputs/CSV/1024bare.1lr.csv")
bare256 = pd.read_csv("./Outputs/CSV/256bare.1lr.csv")
bare256Scheduler = pd.read_csv("./Outputs/CSV/256bareScheduler.csv") # Uses different loss "scale"
bare256SchedulerII = pd.read_csv("./Outputs/CSV/256bareSchedulerMk2.csv")

Adversarial128 = pd.read_csv("./Outputs/CSV/128Adversarial.csv")
Adversarial256 = pd.read_csv("./Outputs/CSV/256Adversarial.csv")
Adversarial64 = pd.read_csv("./Outputs/CSV/64Adversarial.csv")
Adversarial100lr01 = pd.read_csv("./Outputs/CSV/100Adversarial.01.csv")
Adversarial100lr001 = pd.read_csv("./Outputs/CSV/100Adversarial.001.csv")

```

```python
sns.set_style('dark')
sns.set_context("notebook")
bareTest = sns.lineplot(data=bare1024lr1.drop(["Epoch", "Avg_Train_lossA", "Avg_Train_lossB"], axis=1))
bareTest.set(xlabel='Epoch', ylabel='MSE Loss')
# bareTest.figure.savefig("bareTest")
```

```python
sns.set_style('dark')
sns.set_context("notebook")
bareTrain = sns.lineplot(data=bare1024lr1.drop(["Epoch", "TestLossA", "TestLossB"], axis=1))
bareTrain.set(xlabel='Epoch', ylabel='MSE Loss')
# bareTrain.figure.savefig("bareTrain")
```

# Make Comparisons

```python
# dataset1 = 
# dataset2 = 
variable = 'Avg_Train_lossA'
concatenated = pd.concat([bare1024lr01.assign(dataset='1024Latent,.01lr'), bare1024.assign(dataset='1024Latent,.001lr'), bare256Scheduler.assign(dataset='256Latent,Scheduler')])
plot = sns.lineplot(x='Epoch', y=variable, data=concatenated, hue="dataset", palette=sns.color_palette("Set2"))
concatenated.head()
# plot.figure.savefig("Scheduler vs Other LR")
```

```python
variable = 'Avg_Train_lossA' #'TestLossA'
concatenated = pd.concat([bare1024.assign(dataset='1024Latent'), bare2048.assign(dataset='2048Latent'), bare256.assign(dataset='256Latent')])
plot = sns.lineplot(x='Epoch', y=variable, data=concatenated, hue="dataset", palette=sns.color_palette("Set2"))
concatenated.head()
# plot.figure.savefig("Compare perfomance of single encoder models with varying latent space + 256 - Test Losses")
```

```python
variable = 'Avg_Train_lossA' #'Avg_Train_lossA'
concatenated = pd.concat([Adversarial128.assign(dataset='Adversarial128'), Adversarial128.assign(dataset='Adversarial256'), Adversarial64.assign(dataset='Adversarial64')])
plot = sns.lineplot(x='Epoch', y=variable, data=concatenated, hue="dataset", palette=sns.color_palette("Set2"))
concatenated.head()
# plot.figure.savefig("Adversarials vs BareSchedulerII - Test")
```

```python
variable = 'TestLossA' #'Avg_Train_lossA'
concatenated = pd.concat([Adversarial128.assign(dataset='Adversarial128'), bare256SchedulerII.assign(dataset='256BareScheduler')])
plot = sns.lineplot(x='Epoch', y=variable, data=concatenated, hue="dataset", palette=sns.color_palette("Set2"))
concatenated.head()
```

```python
variable = 'TestLossA' #'Avg_Train_lossA'
variable = 'DiscrimA'
concatenated = pd.concat([Adversarial100lr001.assign(dataset='AdversarialLr .001'), Adversarial100lr01.assign(dataset='AdversarialLr .01')])
plot = sns.lineplot(x='Epoch', y=variable, data=concatenated, hue="dataset", palette=sns.color_palette("Set2"))
concatenated.head()
plot.figure.savefig("Adversarial compare LR - Discrim loss")
```
